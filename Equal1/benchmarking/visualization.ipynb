```python
def plot_benchmark(df):
    fig, axes = plt.subplots(1, 3, figsize=(36, 8))

    heatmap_gates_data = df.pivot(index="benchmark", columns="qubits", values="gates")
    sns.heatmap(heatmap_gates_data, annot=True, fmt="d", cmap="YlGnBu", ax=axes[0])
    axes[0].set_title("Heatmap of number of gates for each benchmark")

    heatmap_depth_data = df.pivot(index="benchmark", columns="qubits", values="depth")
    sns.heatmap(heatmap_depth_data, annot=True, fmt="d", cmap="YlGnBu", ax=axes[1])
    axes[1].set_title("Heatmap of depth for each benchmark")

    heatmap_time_data = df.pivot(
        index="benchmark", columns="qubits", values="sample_time"
    )
    heatmap_time_data = heatmap_time_data * 1000
    heatmap_time_data = heatmap_time_data.astype(int)

    sns.heatmap(heatmap_time_data, annot=True, fmt="d", cmap="YlGnBu", ax=axes[2])
    axes[2].set_title("Heatmap of execution time(ms) for each benchmark")

    plt.show()
```

```python
plot_benchmark(df)
```
